```python
import numpy as np

from sklearn.datasets import load_iris
from sklearn.datasets.samples_generator import *
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans as KM
from sklearn.mixture import GaussianMixture as GMM
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
#グラフに色をつけるやつ
def target_to_color(target):
    if type(target) == np.ndarray:
        return (target[0], target[1], target[2])
    else:
        return "rgb"[target]
```

```python
#グラフを出力するやつ
def plot_data(data, target, caption):
    plt.figure()
    plt.scatter(data[:,0], data[:,1], c=[target_to_color(t) for t in target])
    plt.title(caption)
    plt.show()
```

```python
#アヤメデータを二次元にするやつ
def gen_data():
    iris = load_iris()
    pca = PCA(n_components=2)
    return pca.fit_transform(iris.data), iris.target
```

```python
#irisのデータをクラスタリング
data, target = gen_data()
plot_data(data, target, 'iris:original')

km = KM(n_clusters=3)
km_target = km.fit_predict(data)
plot_data(data, km_target, 'iris:k-means')

gmm = GMM(n_components=3, max_iter=1000)
gmm.fit(data)
gmm_target = gmm.predict(data)
plot_data(data, gmm_target, 'iris:GMM')
```

```python
#ブロブデータをクラスタリング
data, target = make_blobs(n_samples=150, n_features=2, centers=3)
plot_data(data, target, 'blobs:original')

km = KM(n_clusters=3)
km_target = km.fit_predict(data)
plot_data(data, km_target, 'blobs:k-means')

gmm = GMM(n_components=3, max_iter=1000)
gmm.fit(data)
gmm_target = gmm.predict(data)
plot_data(data, gmm_target, 'blobs:GMM')
```

```python
#密度の違うブロブデータをクラスタリング
data, target = make_blobs(n_samples=150, cluster_std=[1.0, 2.5, 0.5], random_state=170)
plot_data(data, target, 'blobs2:original')

km = KM(n_clusters=3)
km_target = km.fit_predict(data)
plot_data(data, km_target, 'blobs2:k-means')

gmm = GMM(n_components=3, max_iter=1000)
gmm.fit(data)
gmm_target = gmm.predict(data)
plot_data(data, gmm_target, 'blobs2:GMM')
```
