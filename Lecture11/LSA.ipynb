```python
#手元に音の実験のデータが有ったので試しに使ってみました

from sklearn.decomposition import TruncatedSVD
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 入力データ(A行列にあたる)
A = pd.read_csv('15816054_tachikawa.csv',index_col=0)
A.head()
```

```python
from scipy.linalg import svd

#特異値分解を用いてAを行列成分に分解
U,sigma,VT=svd(A)

# U行列
pd.DataFrame(U).head()
```

```python
# Σ行列
pd.DataFrame(sigma).head()
```

```python
#V^T　行列
pd.DataFrame(VT).head()
```

```python
# Truncated (切り捨てるという意味)SVDで特異値分解と次元削減を行う
# n_componentsは削減後の次元数を表す
svd=TruncatedSVD(n_components=2)
svd.fit(A)
print(svd.transform(A))
```

```python
# 寄与率
print(svd.explained_variance_ratio_)
```

```python
# 累積寄与率
print(sum(svd.explained_variance_ratio_))
```
