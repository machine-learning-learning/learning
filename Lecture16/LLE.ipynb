```python
from sklearn.decomposition import PCA
from sklearn.datasets import samples_generator
from sklearn.manifold import LocallyLinearEmbedding
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cm

#サンプル数
n_samples = 2000

#スイスロールのデータ生成
data, color = samples_generator.make_swiss_roll(n_samples=n_samples)

#表示
fig = plt.figure()
ax = Axes3D(fig)
for i in range(n_samples):
    ax.plot(data[i:i+1,0], data[i:i+1,1], data[i:i+1,2], "o", color=cm.hsv(color[i]/max(color)), ms=4, mew=0.5)
plt.show()
```

```python
#いくつの近傍点を使うか
n_neighbors = 14

#何次元に圧縮するか
n_components = 2

model = LocallyLinearEmbedding(n_neighbors=n_neighbors, n_components=n_components)
model.fit(data)

result = model.transform(data)
for i in range(n_samples):
    plt.scatter(result[i:i+1,0],result[i:i+1,1],color=cm.hsv(color[i]/max(color)))

plt.show()
```
