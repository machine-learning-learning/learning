# 主成分分析

## データの確認

```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
sns.set_style("darkgrid")
cmap = plt.get_cmap("tab10")

 
# 10次元正規乱数を4クラス、合計1000個生成
data_dim = 10
data_num = 500
data_sep = 4

values = np.empty((0,data_dim),float)
sepalated =  data_num // data_sep
for i in range(data_sep):
    cov = np.random.normal(math.log(np.random.rand()*1000)*math.sqrt(i**7)*20, 3, (data_dim, data_dim*100))
    sigma = np.cov(cov, rowvar=1, bias=1)
    values = np.block([[values],[np.random.multivariate_normal(np.random.rand(data_dim)*10, sigma, sepalated)
#                                  + np.array([
#                                      (data_sep-i)*3, 0, np.random.rand(),6*i, [9,-5,8,-1][i],
#                                      i%data_sep*i, [-2,-7,4,8][i], [-10,1,2,1.5][i], 1, i*i/data_sep])
                                ]])
    
# DataFrameの作成
df = pd.DataFrame({ 'Class' : [i//sepalated for i in range(data_num)],
                    'F1' : values[:,0],
                    'F2' : values[:,1],
                    'F3' : values[:,2],
                    'F4' : values[:,3],
                    'F5' : values[:,4],
                    'F6' : values[:,5],
                    'F7' : values[:,6],
                    'F8' : values[:,7],
                    'F9' : values[:,8],
                    'F10' : values[:,9],
                  })
df.head()
```

## 散布図行列を確認

```python
pd.plotting.scatter_matrix(df.iloc[:, 1:], figsize=(8, 8), c=[cmap(i) for i in df.iloc[:, 0]], alpha=0.5)
plt.show()
```

## データの正規化

```python
dfs = df.iloc[:, 1:].apply(lambda x: (x-x.mean())/x.std(), axis=0)
dfs.head()
```

## 分析

```python
from sklearn.decomposition import PCA

# 主成分分析をする
pca = PCA()
pca.fit(dfs)

# データを主成分空間に写像
feature = pca.transform(values)

# 主成分得点
pd.DataFrame(feature, columns=["PC{}".format(x + 1) for x in range(len(dfs.columns))]).head()
```

```python
# 第一主成分でプロットする
fig = plt.figure()
np.histogram(feature[:, 0])
plt.hist(
    [feature[i*sepalated:(i+1)*sepalated, 0] for i in range(data_sep)],
    histtype = "step", 
# stacked = True,
    bins = 100,
    color = [cmap(i) for i in range(data_sep)])
plt.xlabel("PC1")
plt.show()

# 第１主成分と第２主成分でプロットする
fig = plt.figure()
plt.scatter(feature[:, 0], feature[:, 1], color=[cmap(i//sepalated) for i in range(data_num)])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()

# 第１主成分と第２主成分と第３主成分でプロットする
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(feature[:, 0], feature[:, 1],feature[:, 2], color=[cmap(i//sepalated) for i in range(data_num)])
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
plt.show()
```

## 寄与率と累積寄与率

```python
# 寄与率
pd.DataFrame(pca.explained_variance_ratio_, index=["PC{}".format(x + 1) for x in range(len(dfs.columns))])
```

```python
# 累積寄与率
import matplotlib.ticker as ticker
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))
plt.plot([0] + list( np.cumsum(pca.explained_variance_ratio_)), "-o")
plt.xlabel("Number of principal components")
plt.ylabel("Cumulative contribution rate")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
for x, y, name in zip(pca.components_[0], pca.components_[1], df.columns[1:]):
    plt.text(x, y, name)
plt.scatter(pca.components_[0], pca.components_[1], alpha=0.8)
plt.grid()
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
