```python
from sklearn.datasets import load_digits # プチMNIST
import matplotlib.pyplot as plt

data = load_digits()

# 見てみる
plt.axis("off") 
for i in range(36):
    plt.subplot(6, 6, i+1)
    plt.gca().tick_params(labelbottom=0,bottom=0,labelleft=0,left=0)
    plt.imshow(data.images[i], cmap=plt.get_cmap('gray'), interpolation='NONE')
plt.show()
```

### 従来手法での次元削減（PCA）

```python
from sklearn.decomposition import PCA

# 主成分分析をする
pca = PCA(n_components=10)
feature = pca.fit_transform(data.data)

plt.scatter(feature[:, 0], feature[:, 1], color=[cmap(i) for i in data.target])
plt.show()
```

### t-SNE

```python
from sklearn.manifold import TSNE

# 学習
model = TSNE()
result = model.fit_transform(data.data)
```

```python
# 結果を見てみる

cmap = plt.get_cmap("tab10")
for c in range(10):
    plt.scatter([result[i,0] for i in range(len(result)) if data.target[i]==c],
                [result[i,1] for i in range(len(result)) if data.target[i]==c],
                c=[cmap(c) for i in range(len(result)) if data.target[i]==c], label = c)
plt.legend(loc = "upper left", bbox_to_anchor=(1,1))
plt.show()
```

### PCA + t-SNE

```python
model = TSNE(init="pca") # 差分 is ここだけ
result = model.fit_transform(data.data)

cmap = plt.get_cmap("tab10")
for c in range(10):
    plt.scatter([result[i,0] for i in range(len(result)) if data.target[i]==c],
                [result[i,1] for i in range(len(result)) if data.target[i]==c],
                c=[cmap(c) for i in range(len(result)) if data.target[i]==c], label = c)
plt.legend(loc = "upper left", bbox_to_anchor=(1,1))
plt.show()
```
